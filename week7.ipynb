```python
import numpy as np
import pandas as pd

# load data
data1 = pd.read_csv("Cluster datasert 3.csv").values
data2 = pd.read_csv("Cluster datasert 4.csv").values

def kmeans(X, k=2, max_iter=100):
    # random centroids
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(max_iter):
        # assign clusters
        distances = np.linalg.norm(X[:, None] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)

        # update centroids
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])

        # stop if no change
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids

# run on dataset 1
labels1, centroids1 = kmeans(data1, k=2)
print("Dataset 1 centroids:\n", centroids1)

# run on dataset 2
labels2, centroids2 = kmeans(data2, k=2)
print("Dataset 2 centroids:\n", centroids2)
```

```python
import matplotlib.pyplot as plt

def plot_clusters(X, labels, centroids, title):
    plt.scatter(X[:,0], X[:,1], c=labels)
    plt.scatter(centroids[:,0], centroids[:,1], c='red', marker='X')
    plt.title(title)
    plt.show()

plot_clusters(data1, labels1, centroids1, "Dataset 1")
plot_clusters(data2, labels2, centroids2, "Dataset 2")
```

```python
from sklearn.cluster import KMeans
import pandas as pd

data = pd.read_csv("Cluster datasert 3.csv").values

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

print(kmeans.cluster_centers_)
print(kmeans.labels_)
```
