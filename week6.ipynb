```python
import numpy as np

# sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# training function (1 epoch)
def train(X, Y, w, b, lr=0.5):
    for i in range(len(X)):
        x = X[i]
        y = Y[i]

        # forward
        z = np.dot(w, x) + b
        y_hat = sigmoid(z)

        # error
        error = y - y_hat

        # gradient
        delta = error * y_hat * (1 - y_hat)

        # update
        w += lr * delta * x
        b += lr * delta

        print(f"x={x}, y={y}, y_hat={round(y_hat,3)}, w={w}, b={round(b,3)}")

    return w, b


# ---------------- DATASETS ---------------- #

# Dataset 1 (OR)
X1 = np.array([[0,0],[0,1],[1,0],[1,1]])
Y1 = np.array([0,1,1,1])

# Dataset 2 (AND)
X2 = np.array([[0,0],[0,1],[1,0],[1,1]])
Y2 = np.array([0,0,0,1])

# Dataset 3 (continuous)
X3 = np.array([[0,0],[0.5,0.5],[1.3,0.5],[0.8,1.2]])
Y3 = np.array([0,1.1,1.2,1.0])


# initial weights
w = np.array([0.5, 0.6], dtype=float)
b = 0.3

print("\n--- Dataset 1 ---")
w, b = train(X1, Y1, w, b)

print("\n--- Dataset 2 ---")
w, b = train(X2, Y2, w, b)

print("\n--- Dataset 3 ---")
w, b = train(X3, Y3, w, b)
```
